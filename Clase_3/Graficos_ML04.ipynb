```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x= [1,2,3,4,5,6,7,8]
y= [5,2,4,2,1,4,5,2]
```

GRAFICO DE LINEAS

```python
plt.plot(x,y)
plt.title("Grafico de Lineas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.show()
```

GRAFICO DE BARRAS

```python
plt.bar(x,y, color='red', width=0.8, align='center')
plt.title("Grafico de Lineas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.show()
```

GRAFICO DE PARETTO

```python
categorias= ['A', 'B', 'C', 'D', 'E']
frecuencias = [50,30,10,7,3]
orden = np.argsort(frecuencias)[::-1]
categorias_ordenadas = [categorias[i] for i in orden]
frecuencias_ordenadas = [frecuencias[i] for i in orden]
```

```python
categorias, frecuencias
```

```python
acumulado = np.cumsum(frecuencias_ordenadas)
acumulado_porcentaje = acumulado / np.sum(frecuencias_ordenadas) * 100
```

```python
acumulado, acumulado_porcentaje
```

```python
fig, ax = plt.subplots()
ax.bar(categorias_ordenadas, frecuencias_ordenadas, color='blue')
ax.set_xlabel('Categorias')
ax.set_ylabel('Frecuencia')
ax.set_title('Grafico de Pareto')
ax2 = ax.twinx()
ax2.plot(categorias_ordenadas, acumulado_porcentaje, color='red', marker='o')
ax2.set_ylabel('Porcentaje Acumulado')
plt.show()
#
```

```python
import seaborn as sns
```

```python
# Datos con categorías adicionales
data = pd.DataFrame({
    "Categoria": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Subgrupo": ["X", "Y"] * 4,
    "Valor": [10, 12, 15, 17, 7, 8, 12, 14]
})

# Gráfico con hue para colorear por subgrupo
sns.barplot(x="Categoria", y="Valor", hue="Subgrupo", data=data)

plt.title("Ejemplo con Seaborn - hue")
plt.show()
```

```python
x= np.linspace(0,2*np.pi, 100)
```

```python
fig, axs= plt.subplots(2,2, figsize=(10,10))
axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Seno")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Coseno")

axs[1,0].plot(x, np.tan(x))
axs[1,0].set_title("Tangente")

axs[1,1].plot(x, np.arctan(x))
axs[1,1].set_title("Arcotangente")

plt.tight_layout()
plt.show()
```
